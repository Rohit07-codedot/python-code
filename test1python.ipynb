```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    
    if x not in seen:
        seen.add(x)
        
    return list(seen)
```

```python
from typing import List

def moving_average(stream: List[int], w: int) -> List[float]:
    result = []
    window_sum = 0
    
    for i in range(len(stream)):
        window_sum += stream[i]
        
       
        if i >= w:
            window_sum -= stream[i - w]
        
        
        current_window_size = min(i + 1, w)
        
        result.append(window_sum / current_window_size)
    
    return result

    print(moving_average(stream, w))
```

```python
def expand_string(s: str) -> str:
    result = []
    
    for i in range(0, len(s), 2):
        count = int(s[i])
        char = s[i + 1]
        result.append(char * count)
    
    return "".join(result)



print(expand_string("3a2b1c"))
```
